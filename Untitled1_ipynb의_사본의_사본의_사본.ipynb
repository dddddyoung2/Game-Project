2) 특이값 처리
- Platform : '2600'
 - Game Platform 중 Atari 2600 비디오 컴퓨터 시스템이 실제 존재함. 처리 안함
- Year : 4자리 아닌 2자리 존재
 - 만약 x가 20보다 작으면 2000 더하기
 - 만약 x가 20보다 20크고 99보다 작으면 1900 더하기

- Sales : M, K
 - M: 'M'제거
 - K: 'K'제거 후 나누기 1000

````python
#Year 처리
def func(x):
  if x < 20:
    return x+2000
  elif 20 < x < 99 :
    return x+1900
  else:
    return x

df2['Year']=df2['Year'].apply(func)
df2['Year']=df2['Year'].astype(int)
````

````python
#Sales 처리_M

def change_M(x):
  try:
    return x.replace('M','')                                                                                                                                                      
  except:
    return x

df2['NA_Sales']=df2['NA_Sales'].apply(change_M)
df2['EU_Sales']=df2['EU_Sales'].apply(change_M)
df2['JP_Sales']=df2['JP_Sales'].apply(change_M)
df2['Other_Sales']=df2['Other_Sales'].apply(change_M)
````

````python
#Sales 처리_K_살짝 NGD

def change_K(x):
  try:
    return float(x.replace('K','')) #'K' 제거 후 실수처리
  except:
    return float(x)

df2['NA_Sales']=df2['NA_Sales'].apply(change_K)
df2['EU_Sales']=df2['EU_Sales'].apply(change_K)
df2['JP_Sales']=df2['JP_Sales'].apply(change_K)
df2['Other_Sales']=df2['Other_Sales'].apply(change_K)
````

````python
#Sales 처리_K_살짝 NGD_2

def change_K2(x):
  if x==10 or x==20 or x==30 or x==40 :
    return x/1000
  elif x>=50 :
    return x/1000
  else:
    return x

df2['NA_Sales']=df2['NA_Sales'].apply(change_K2)
df2['EU_Sales']=df2['EU_Sales'].apply(change_K2)
df2['JP_Sales']=df2['JP_Sales'].apply(change_K2)
df2['Other_Sales']=df2['Other_Sales'].apply(change_K2)

df2.to_csv('df2.csv')
````

````python
#국가별 출고량 차이 검정

unique_NA=pd.DataFrame(df2.groupby('Genre')['NA_Sales'].sum())
unique_EU=pd.DataFrame(df2.groupby('Genre')['EU_Sales'].sum())
unique_JP=pd.DataFrame(df2.groupby('Genre')['JP_Sales'].sum())
unique_Other=pd.DataFrame(df2.groupby('Genre')['Other_Sales'].sum())
unique_Global=pd.DataFrame(df2.groupby('Genre')['Global_Sales'].sum())
````

````python
#국가별 장르 출고량 df
df3=pd.concat([unique_NA,unique_EU,unique_JP,unique_Other, unique_Global], join='inner', axis=1) #merge
df3=df3.sort_values('Global_Sales', ascending=False) #내림차순
df3
````

````python
#분산 분석(ANOVA)
#3개 이상분산 분석(ANOVA)
#3개 이상의 집단의 평균이 같은지/차이가 있는지 확인
#H0: 모든 집단의 평균이 같다.

from scipy.stats import f_oneway
fstat, pval = f_oneway(df3['NA_Sales'], df3['EU_Sales'],df3['JP_Sales'],df3['Other_Sales'])

print('국가별 장르 출고량 분산 분석(P-value): ', pval)
````

- ANOVA 분석 결과, p<0.05로 귀무가설 기각 대립가설 채택

- 국가별 장르의 출고량 평균을 비교 해봄

- '지역에 따라서 선호하는 게임 장르는 차이가 있다' -> 어떤 집단에서 차이가 있는지 알아보기 위해 사후 검정 실시

````python
#사후 검정
#Tukey's Range Test

북미=df3['NA_Sales']
유럽=df3['EU_Sales']
일본=df3['JP_Sales']
기타=df3['Other_Sales']

from scipy.stats import f_oneway
import numpy as np
from statsmodels.stats.multicomp import pairwise_tukeyhsd
fstat, pval = f_oneway(북미, 유럽, 일본, 기타)

# np.concatenate를 통해 데이터를 합쳐 놓는다.
v = np.concatenate([북미, 유럽, 일본, 기타])


# 데이터 개수만큼 레이블을 준비한다.
labels = ['북미']*len(북미) + ['유럽']*len(유럽) + ['일본']*len(일본) + ['기타']*len(기타)
#labels = [df3['NA_Sales']]*len(df3['NA_Sales']) + [df3['EU_Sales']*len(df3['EU_Sales']) + [df3['JP_Sales']]*len(df3['JP_Sales']) + [df3['Other_Sales']]*len(df3['Other_Sales'])


# 사후 검정을 수행한다.
tukey_results = pairwise_tukeyhsd(v, labels, 0.05)
print(tukey_results)
````

- 국가별 사후검정 결과(어떤  집단이 차이가 있는지 확인하기 위해)

```
data1: NA_Sales(북미)
data2: EU_Sales(유럽)
data3: JP_Sales(일본)
data4: Other_Sales(기타)
```



- reject 컬럼에서 True로 적힌 data1 & data3과 data1 & data4 는 평균이 다르다.
- 선호하는 장르가 다르다고 할 수 있다.


```
data1 & data3: 북미 출고량 & 일본 출고량
data1 & data4: 북미 출고량 & 기타 국가 출고량
```




- 전세계 장르 출고량 차이 비교 결과, p(0.0)<0.05로 귀무가설 기각
- 결론: 장르별로 출고량은 차이가 있다.
